```python
def porcentagem(variavel):
    tamanho = variavel.sum()

    soma = valores = 0
    
    print()

    for i in variavel * 100 / tamanho:
        print(f'{variavel.index[valores]:20} --> {variavel[valores]:^3} --> {i:.0f}%')
        soma += i
        valores += 1

    print('-' * 20)
    print(f'Total: {soma}%')
    print(f'{tamanho} peças')
```
