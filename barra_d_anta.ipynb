# Definição de matérias-primas

````python
df_silexito = df[df['Matéria-prima'] == 'Silexito']
df_indeterminado = df[df['Matéria-prima'] == 'Indeterminado']
df_quartzo_leitoso = df[df['Matéria-prima'] == 'Quartzo leitoso']
df_quartzito = df[df['Matéria-prima'] == 'Quartzito']
df_arenito_silicificado = df[df['Matéria-prima'] == 'Arenito silicificado']
df_basalto = df[df['Matéria-prima'] == 'Basalto']
df_quartzo_hialino = df[df['Matéria-prima'] == 'Quartzo hialino']
````

# Distribuição de matérias-primas

````python
# Tirei o quartzo hialino, pois só tinha uma peça e nem aparecia no gráfico
fig, (ax1) = plt.subplots(nrows = 1, figsize = (10,6))

ax1.spines['top'].set_color('black'), ax1.spines['top'].set_visible(True), ax1.spines['top'].set_linewidth(1)
ax1.spines['bottom'].set_color('black'), ax1.spines['bottom'].set_visible(True), ax1.spines['bottom'].set_linewidth(1)
ax1.grid(True, color = 'black', linestyle = '-', linewidth = 0.1, alpha = 0.4), ax1.set_facecolor('white')

df['Matéria-prima'].value_counts().plot.barh(color = lightblue)

plt.xlabel('Quantidade')
plt.title('Distribuição de matérias-primas');
````

# Localização das coletas

````python
fig, (ax1) = plt.subplots(nrows = 1, figsize = (8,5))

ax1.spines['top'].set_color('black'), ax1.spines['top'].set_visible(True), ax1.spines['top'].set_linewidth(1)
ax1.spines['bottom'].set_color('black'), ax1.spines['bottom'].set_visible(True), ax1.spines['bottom'].set_linewidth(1)
ax1.grid(True, color = 'black', linestyle = '-', linewidth = 0.1, alpha = 0.4), ax1.set_facecolor('white')

df[df['Coleta'] != '-']['Coleta'].value_counts().plot.barh(color = azul, alpha = 0.6)

plt.xlabel('Quantidade')
plt.title('Distribuição dos materiais nas áreas de coleta');
````

````python
def porcentagem(variavel):
    tamanho = variavel.sum()

    soma = valores = 0
    
    print('Dados de percentagens:')
    print()

    for i in variavel * 100 / tamanho:
        print(f'{variavel.index[valores]:20}--> {variavel[valores]:<3} peças --> {i:.0f}%')
        soma += i
        valores += 1

    print()
    print(f'Total: {soma}%')
````
